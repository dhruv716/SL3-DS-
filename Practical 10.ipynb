```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/Users/dhruvpai/Downloads/IRIS.csv")
sns.histplot(df['sepal_length'],kde=False,bins=10)
plt.show()
sns.histplot(df['sepal_width'],kde=False,bins=10)
plt.show()
sns.histplot(df['petal_length'],kde=False,bins=10)
plt.show()
sns.histplot(df['petal_width'],kde=False,bins=10)
plt.show()
```

```python
sns.histplot(df['sepal_length'],kde=True,bins=10)
plt.show()
sns.histplot(df['sepal_width'],kde=True,bins=10)
plt.show()
sns.histplot(df['petal_length'],kde=True,bins=10)
plt.show()
sns.histplot(df['petal_width'],kde=True,bins=10)
plt.show()
```

```python
sns.scatterplot(df['sepal_length'])
plt.show()
sns.scatterplot(df['sepal_width'])
plt.show()
sns.scatterplot(df['petal_length'])
plt.show()
sns.scatterplot(df['petal_width'])
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df) 
plt.show()
sns.boxplot(x='species', y='sepal_width', data=df) 
plt.show()
sns.boxplot(x='species', y='petal_length', data=df) 
plt.show()
sns.boxplot(x='species', y='petal_width', data=df) 
plt.show()
sns.pairplot(df,hue='species')
plt.show()
```
